```python
import math
import string
import nltk
string.punctuation
#from __future__ import print_function

stop_list=nltk.corpus.stopwords.words('english')

suffix_token=[]
prefix_token=[]
prefix_dict={}

def rem_punctuation(text):
    txt_np="".join([c for c in text if c not in string.punctuation])
    return txt_np


class Perform_operation():
    
    @staticmethod
    def List_AND_operation(Op_left, Op_right):
        # perform 'merge'
        
        pos_right = 0        # current position in right_operand
        pos_left = 0          # current position in left_operand
        
        answer = []           # answer list to be returned
        
        x=len(Op_right)
        y=len(Op_left)
        jump_r = int(math.sqrt(x)) # jump pointer distance for r_index
        jump_l = int(math.sqrt(y))  # jump pointer distance for l_index

        while (pos_left < y and pos_right < x):
            item_right = Op_right[pos_right]          # current item in right_operand
            
            item_left = Op_left[pos_left]             # current item in left_operand
             
            
           
            if (item_left == item_right):
                answer.append(item_left)   
                
                pos_right = pos_right +1            # advance right position
                pos_left = pos_left+ 1            # advance left position
            
            
            elif (item_left > item_right):             # case : if left item is more than right item
                
                aa=pos_right +jump_r
            
                if (aa < x) and Op_right[aa] <= item_left:
                    pos_right = pos_right+ jump_r
                # else advance r_index by 1
                else:
                    pos_right =pos_right+ 1

            
            else:                                        # case : if left item is less than right item
               
                bb=pos_left + jump_l
                if ( bb< y) and Op_left[bb] <= item_right:
                    pos_left = pos_left +jump_l
                
                else:
                    pos_left = pos_left+1

        return answer

    @staticmethod
    def List_OR_operation(Op_left, Op_right):
             # union of left and right operand
        
        answer = []    
        pos_right = 0     # current position in right_operand
        pos_left = 0      # current position in left_operand
        x=len(Op_right)
        y=len(Op_left)
        
        while (pos_left < y or pos_right < x):
           
            if (pos_left < y and pos_right < x):
                item_left = Op_left[pos_left]                # current item in left_operand
                item_right = Op_right[pos_right]             # current item in right_operand
                
                # case : if items are equal, add one of them to result 
                if (item_left == item_right):
                    answer.append(item_left)
                    pos_left = pos_left+1
                    pos_right =pos_right+ 1

                # case : left item greater than right item
                elif (item_left > item_right):
                    answer.append(item_right)
                    pos_right =pos_right+ 1

                # case : left item lower than right item
                else:
                    answer.append(item_left)
                    pos_left = pos_left+1

            # if left operand list is over, append right item move ahead
            elif (pos_left >= y):
                item_right = Op_right[pos_right]
                answer.append(item_right)
                pos_right =pos_right+ 1

            #  if right operand list is over, append left item move ahead
            else:
                item_left = Op_left[pos_left]
                answer.append(item_left)
                pos_left =pos_left+ 1

        return answer

    @staticmethod
    def List_NOT_operation(Op_right, Doc_Inv_ind):
       
        if (not Op_right):
            return Doc_Inv_ind
        
        answer = []
        pos_right = 0 
        for item in Doc_Inv_ind:
             
            if (item != Op_right[pos_right]):
                answer.append(item)
           
            elif (pos_right + 1 < len(Op_right)):
                pos_right =pos_right+ 1
        return answer
```

```python


class Node(object):
    
    def __init__(self, value):
        
        self.left = None
        self.right = None
        self.value = value

    def insert(self, value):
        
        if self.value:
            if value < self.value:
                if self.left is None:
                    self.left = Node(value)
                else:
                    self.left.insert(value)
            elif value > self.value:
                if self.right is None:
                    self.right = Node(value)
                else:
                    self.right.insert(value)
        else:
            self.value = value
    

    def traversal(self):
        
        #traverse by stack 
        tre_stack = []
        node = self
        while tre_stack or node:
            if node:
                tre_stack.append(node)
                node = node.left
            else:
               
                node = tre_stack.pop()
                yield node.value
                node = node.right

    
```

```python
import nltk
import collections

#from boolean import BooleanModel
hold1=[]
hold2=[]
hold3=[]


class Info_Retrieval():

    def __init__(self, docs=None, stop_words=stop_list):
        if docs is None:
            raise UserWarning('Zero Docs')
        self.documents = docs
        self.stemming = nltk.stem.porter.PorterStemmer()
        self.Inverted_index = self.Op_preprocessing(stop_words)
        self.Inverted_index1 = self.preproc(stop_words)
        self.Print_Id()
        #self.tri()
    
    
    def preproc(self,stop_words=stop_list):
        fr_index = {}
        for ptr, doc in enumerate(self.documents):
            for word in doc.split():
                token=rem_punctuation(word.lower()) 
                hold3.append(token)
                if ((token not in stop_words) and (len(token)<40) and token.isnumeric()==False):
                    if fr_index.get(token, -244) == -244:
                        fr_index[token] = Node(ptr + 1)
                    elif isinstance(fr_index[token], Node):
                        fr_index[token].insert(ptr + 1)
                    else:
                        raise UserWarning('error for posting list')
        print(hold3[0:20])            
        return fr_index
    
    
    def Op_preprocessing(self,stop_words=stop_list):
        list_index = {}
        for ptr, doc in enumerate(self.documents):
            for word in doc.split():
                
                
                hold2.append(word)
                token=rem_punctuation(word) 
                token = self.stemming.stem(token.lower())
                if ((token not in stop_words) and (len(token)<40) and token.isnumeric()==False): 
                    hold1.append(token)
                    if list_index.get(token, -244) == -244:
                        list_index[token] = Node(ptr + 1)
                    elif isinstance(list_index[token], Node):
                        list_index[token].insert(ptr + 1)
                    else:
                        raise UserWarning('error for posting list')
        print(hold2[0:20])
        print(hold1[0:20])
        return list_index
    
    
        
       
   
    def Print_Id(self):
        print('Inverted_Index:\n')
        for word, tree in self.Inverted_index1.items():
            prefix_token.append(word)
            prefix_dict[word]=len([num_doc for num_doc in tree.traversal() if num_doc != None])
            print('{}: {}'.format(word, [num_doc for num_doc in tree.traversal() if num_doc != None]))
        print()
        
        for word, tree in self.Inverted_index.items():
            suffix_token.append(word)
            print('{}: {}'.format(word, [num_doc for num_doc in tree.traversal() if num_doc != None]))
        print()

    def Posting_list(self, word):
        return [num_doc for num_doc in self.Inverted_index[word].traversal() if num_doc != None]

    @staticmethod
    def Query_parsing(infix_tokens):
        """ Query_parsing
         Shunting Yard Algorithm used
        """
        Priority = {}
        Priority['('] = 0
        Priority[')'] = 0  
        Priority['NOT'] = 3
        Priority['AND'] = 2
        Priority['OR'] = 1
         

        queue_out = []
        stack_operator = []
        minus=-1
        st='('
        ed=')'

        for token in infix_tokens:
            if (token == st):
                stack_operator.append(token)
            
            # if right bracket, pop all operators from operator stack onto output until we hit left bracket
            elif (token == ed):
                nxt_op = stack_operator.pop()
                while nxt_op != st:
                    queue_out.append(nxt_op)
                    nxt_op = stack_operator.pop()
            
            # if operator, pop operators from operator stack to queue if they are of higher precedence
            elif (token in Priority):
                
                if (stack_operator):
                    current_operator = stack_operator[minus]
                    
                    while (stack_operator and Priority[current_operator] > Priority[token] ):
                        queue_out.append(stack_operator.pop())
                        if (stack_operator):
                            current_operator = stack_operator[minus]
                stack_operator.append(token) 
            else:
                queue_out.append(token.lower())

        # while there are still operators on the stack, pop them into the queue
        while (stack_operator):
            queue_out.append(stack_operator.pop())

        return queue_out

    def Query_processing(self, User_query):
        # prepare User_query list
        User_query_f = User_query.replace('(', '( ')
        User_query_s = User_query_f.replace(')', ' )')
        User_query = User_query_s.split(' ')
        intersect='AND'
        merge='OR'
        negate='NOT'
        on=1
        
        Doc_Inv_ind = list(range(on, len(self.documents) + on))
        
        get_queue_postfix = collections.deque(self.Query_parsing(User_query)) 
        stack_res = []
       

        while get_queue_postfix:
            token = get_queue_postfix.popleft()
            answer = []                                    # the evaluated result at each stage
           
            if (token != intersect and token != merge and token != negate):
                token = self.stemming.stem(token)                       
                                                                 
                if (token in self.Inverted_index):
                    answer = self.Posting_list(token)
            
            elif (token == 'OR'):                            # merging of list
                Op_right = stack_res.pop()
                Op_left = stack_res.pop()
                answer = Perform_operation.List_OR_operation(Op_left, Op_right) 
                
            elif (token == 'AND'):
                Op_right = stack_res.pop()
                Op_left = stack_res.pop()
                answer = Perform_operation.List_AND_operation(Op_left, Op_right)   #  intersection of lists

            elif (token == 'NOT'):
                Op_right = stack_res.pop()
                answer = Perform_operation.List_NOT_operation(Op_right, Doc_Inv_ind) # negation

            stack_res.append(answer)                        

       
        num=len(stack_res)
        if num != on: 
            print("ERROR: Invalid User_query. Please check User_query syntax.") # error checking
            return None
        
        return stack_res.pop()
    
```

```python
def querying(query):
    final_result=[]
    queryA_list=[]
    queryB_list=[]
    case4inter=[]
    final_output=[]
    #final_output=[]
   
    # Split query and determine it's type
    parts = query.split("*")

    final_result.append('Query Processed as:-')
    final_result.append(parts)
   
    #These are the different cases formed depending on the number of wild card characters and their position in our query
    if len(parts)==1:
        case =0
    elif len(parts) == 3:
        case = 4
    elif parts[1] == "":
        case = 1
    elif parts[0] == "":
        case = 2
    elif parts[0] != "" and parts[1] != "":
        case = 3

    #Case 4 is dealt sperately as it has 2 sub queries    
    if case == 4:
        if parts[0] == "":
            case = 1

    # Read Permuterm Index
    permuterm = {}
    with open("perm.txt") as f:
        for line in f:
            temp = line.split()
            if(len(temp)>1):
                permuterm[temp[0]] = temp[1]
   
    #This function will match the prefix of the word/wildcard query to the words in index
   


    def intersection(lst1, lst2):
        return list(set(lst1) & set(lst2))
   
    def prefix_match(term, prefix):
        term_list = []
        for tk in term.keys():
            if tk.startswith(prefix):
                 #final_result.append(tk)     # Permuterm Index where wildcard query is matched
                term_list.append(term[tk])
        return term_list

    #This function is used to process query (ie after prefix match, the word and document is extracted where the prefix match has occured)
    def process_query(query):    
        term_list = prefix_match(permuterm,query)
        final_result.append('Words Matching Wildcard Query:-')
        final_result.append(term_list)
        return term_list

    #Queries are processed on the bases of their cases
    if case == 0:
        pass
    elif case == 1:
        if (parts[0]==''):
            query = parts[1]
            final_result.append('This is how the query will be processed')
            final_result.append(query)
        else:
            query = "$" + parts[0]
            final_result.append('This is how the query will be processed')
            final_result.append(query)
    elif case == 2:
        query = parts[1] + "$"
        final_result.append('This is how the query will be processed')
        final_result.append(query)
    elif case == 3:
        query = parts[1] + "$" + parts[0]
        final_result.append('This is how the query will be processed')
        final_result.append(query)      
    elif case == 4:
        queryA = parts[2] + "$" + parts[0]
        queryB = parts[1]
        final_result.append('This is how the query will be processed')
        final_result.append([queryA, queryB])

    if case != 4:
        return process_query(query)

    elif case == 4:
       
    # query A: Z$X*
        queryA_list = prefix_match(permuterm,queryA)
        final_result.append('This is out List contating the terms which match our desired queryA')
        final_result.append(queryA_list)
    # query B: Y*
        queryB_list= prefix_match(permuterm,queryB)
        final_result.append('This is out List contating the terms which match our desired queryB')
        final_result.append(queryB_list)  
       
    # Intersection of Query A and Query B words
        queryA_and_queryB = intersection(queryA_list,queryB_list)
        case4inter=queryA_and_queryB
        final_result.append('This is List contating common term documents for queryA and queryB')
        final_result.append(queryA_and_queryB)
        #final_output.append(queryA_and_queryB)
        return case4inter
   
```

```python
from collections import Counter
words = prefix_token
counts = prefix_dict

def correct(word):
    "Find the best spelling correction for this word."
    # Prefer edit distance 0, then 1, then 2; otherwise default to word itself.
    candidates = (known(edits0(word)) or 
                  known(edits1(word)) or 
                  known(edits2(word)) or 
                  [word])
    return max(candidates, key=counts.get)

def known(words):
    "Return the subset of words that are actually in the dictionary."
    return {word for word in words 
                if word in counts}

def edits0(word): 
    "Return all strings that are zero edits away from word (i.e., just word itself)."
    return {word}

def edits2(word):
    "Return all strings that are two edits away from this word."
    return {e2 for e1 in edits1(word) 
                for e2 in edits1(e1)}

def edits1(word):
    "Return all strings that are one edit away from this word."
    pairs      = splits(word)
    deletes    = [a+b[1:]           for (a, b) in pairs if b]
    transposes = [a+b[1]+b[0]+b[2:] for (a, b) in pairs if len(b) > 1]
    replaces   = [a+c+b[1:]         for (a, b) in pairs for c in alphabet if b]
    inserts    = [a+c+b             for (a, b) in pairs for c in alphabet]
    return set(deletes + transposes + replaces + inserts)

def splits(word):
    "Return a list of all possible (first, rest) pairs that comprise word."
    return [(word[:i], word[i:]) 
                for i in range(len(word)+1)]

from string import ascii_lowercase as alphabet

assert alphabet == 'abcdefghijklmnopqrstuvwxyz'

import re
def tokens(text):
    "List all the word tokens (consecutive letters) in a text. Normalize to lowercase."
    return re.findall('[a-z]+', text.lower())

```
